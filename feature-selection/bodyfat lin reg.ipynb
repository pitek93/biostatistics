```r
a=file.choose()
b=read.table(a,header=F)
```

```r
x=b[,3:15]
y=b[,2]
```

```r
# corr y with x
cor(y,x)
```

```r
# linear regression
z=lm(y~.,data=as.data.frame(x))
```

```r
summary(z)
```

```r
# R^2 = 0.75, linear model is well fitted
```

```r
# Akaike Criteria - elimination method
step(z, direction="backward", k=2)
```

```r
# features chosen: 3,4,6,8,9,10,14,15
```

```r
# new features model quality
zA=lm(y~.,data=as.data.frame(x[,c(1,2,4,6,7,8,12,13)]))
```

```r
summary(zA)
```

```r
#R2 = 0.74. Quality didn't change too much (even though we've removed 5 features)
# V9 seems useless as well
```

```r
# Shwartz Criteria (Bayes) - elimination method
step(z,direction="backward", k=log(252))
```

```r
# Elimination keeps 4,8,14,15 features
```

```r
zS=lm(y~.,data=as.data.frame(x[,c(2,6,12,13)]))
```

```r
# model quality
summary(zS)
```

```r
# 9 features removed, quality is almost the same.
```

```r
# Akaike - forward method
# Initialising clean model
z0=lm(y ~ 1, data=as.data.frame(x))
summary(z0)
```

```r
step(z0, k=2, direction="forward", scope=formula(z))
```

```r
# The same variables are kept as with backward (elimination) Akaike method
```

```r
# Bayes Criterion, forward method
step(z0, k=log(252), direction="forward", scope=formula(z))
```

```r
# Results are the same as with Bayes backward method
```

```r
# Akaike criteria - direction both
```

```r
z=lm(y ~., data=as.data.frame(x))
```

```r
step(z, k=2, direction="both")
```

```r
#same results as with previous Akaike
```

```r
# Bayes method - direction both:
```

```r
step(z, k=log(252), direction="both")
```

```r
#same results as previous
#Shwartz picks 4,8,14,15
#values for 14 and 15 dont increase a value too much
```

```r
#removing var 14
z14=lm(y~.,data=as.data.frame(x[,c(2,6,13)]))
summary(z14)
```

```r
#R2=0.72
```

```r
#removing var 15
z15=lm(y~.,data=as.data.frame(x[,c(2,6)]))
summary(z15)
```

```r
# R2=0.71
# small value of Shwartz criteria
```

```r
# removing var 4
z4=lm(y~.,data=as.data.frame(x[,6]))
summary(z4)
```

```r
#R2=0.66 - decrease in quality
```
